```python
import matplotlib.pyplot as plt
import numpy as np
from cmath import exp, pi

def dft(y):
    '''
    This is a code to do decrete Fourier transform
    '''    
    N = len(y)
    c = np.zeros(N, complex)
    for k in range(N):
        for n in range(N):
            c[k] += y[n]*exp(-2j*pi*k*n/N)
    return c

def idft(c):
    '''
    This is a code to do inverse Fourier transform
    '''
    N = len(c)
    y = np.zeros(N, complex)
    for k in range(N):
        for n in range(N):
            y[k] += c[n]*exp(2j*pi*k*n/N)
    return y

#create the initial sample points (x,y)

x = []
y = []

for num in np.loadtxt('years.txt'):
    x.append(num)

for num in np.loadtxt('tropstorm.txt'):
    y.append(num)
    
#adjust size and dpi
fig = plt.figure(figsize=(15, 5), dpi=100)

#performat DFT and inverse dft calculations
c = dft(y)    
y1 = idft(c).real/len(y) 

#title
plt.title("Tropical Storms in Atlantic Since 1851")

#make the plot
plt.plot(x,y, 'b*', label='original function')
plt.plot(x,y1, 'y', label='idft')

plt.xlabel('Year')
plt.ylabel('Number of Storms')
plt.xlim([1851,2016])
plt.legend()
plt.show()
```

```python
c2 = c[np.abs(c)>21]           # remove the small coefficients
y2 = idft(c2).real/len(y)     # performa inverse DFT
x2 = np.linspace(1851,2016,len(y2))  # obtain the x sequence

#adjust size and dpi
fig = plt.figure(figsize=(15, 5), dpi=100)

#title
plt.title("Tropical Storms in Atlantic Since 1851")

plt.plot(x,y, label='original data', linewidth = 1.0, color = 'blue')
plt.plot(x2,y2, 'r--', label='idft', linewidth = 3.0, color = 'red')

plt.xlabel('Year')
plt.ylabel('Number of Storms')
plt.xlim([1851,2016])
plt.legend()
plt.show()
```
