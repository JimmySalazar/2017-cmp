```python
import matplotlib.pyplot as plt
import numpy as np

x = []
y = []

for num in np.loadtxt('years.txt'):
    x.append(num)

for num in np.loadtxt('tropstorm.txt'):
    y.append(num)

#adjust size and dpi
fig = plt.figure(figsize=(15, 5), dpi=100)

#title
plt.title("Tropical Storms in Atlantic Since 1851")

#labels
plt.xlabel("Year")
plt.ylabel("Number of Storms")

#X and Y Axis limits
plt.xlim([1851, 2016])

#fit
fit = np.polyfit(x, y, 5)
y_p = np.polyval(fit, x)

#plot

plt.plot(x, y, label = 'Actual')
plt.plot(x, y_p, label = 'Trend', linewidth = 3.0, color = 'black')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = []
y = []

for num in np.loadtxt('years.txt'):
    x.append(num)

for num in np.loadtxt('ace.txt'):
    y.append(num)

#adjust size and dpi
fig = plt.figure(figsize=(15, 5), dpi=100)

#title
plt.title("Accumulated Cyclone Energy Since 1851")

#labels
plt.xlabel("Year")
plt.ylabel("Acumulated Cyclone Energy")

#X and Y Axis limits
plt.xlim([1851, 2016])

#fit
fit = np.polyfit(x, y, 5)
y_p = np.polyval(fit, x)
#print(sum((y_p-y)**2))

#plot
plt.axhline(y=111, xmin=0, xmax=1, hold = None, color="red")
plt.axhline(y=66, xmin=0, xmax=1, hold = None, color="red")
plt.plot(x, y, label = 'Actual')
plt.plot(x, y_p, label = 'Trend', linewidth = 3.0, color = 'black')
plt.legend()
plt.show()
print("""* Accumulated cyclone energy (ACE) is a measure used by the NOAA to express 
      the activity of individual tropical cyclones and entire tropical cyclone seasons. 
      It uses an approximation of the wind energy used by a tropical system over its 
      lifetime and is calculated every six hours. The ACE of a season is the sum of the 
      ACEs for each storm and takes into account the number, strength, and duration of 
      all the tropical storms in the season. """)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = []
y = []
z = []

for num in np.loadtxt('years.txt'):
    x.append(num)

for num in np.loadtxt('tropstorm.txt'):
    y.append(num)
    
for num in np.loadtxt('hurricane.txt'):
    z.append(num)
    
combined = np.add(y, z)

#adjust size and dpi
fig = plt.figure(figsize=(15, 5), dpi=100)

#title
plt.title("Sum of Tropical Storms & Hurricanes in Atlantic Since 1851")

#labels
plt.xlabel("Year")
plt.ylabel("Number of Storms")

#X and Y Axis limits
plt.xlim([1851, 2016])

#fit
fit = np.polyfit(x, combined, 5)
y_p = np.polyval(fit, x)
#print(sum((y_p-y)**2))

#plot
plt.axhline(y=22, xmin=0, xmax=1, hold = None, color="red")
plt.plot(x, combined, label = 'Actual')
plt.plot(x, y_p, label = 'Trend', linewidth = 3.0, color = 'black')
plt.legend()
plt.show()
```
