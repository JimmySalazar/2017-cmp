# Linear Line Fitting of Streamflow Data from the Colorado River at Lees Ferry (1906-2005)

```python
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
import pandas as pd


# Define Empty List for data

x = [] # Years from 1906 - 2005
y = [] # Actual Measured Data 1906 - 2005
z = [] # Tree Ring Data 1906 - 2005

# Load Data & Append to Empty Lists

for num in np.loadtxt('Years_Observed.txt'):
    x.append(num)
    
for num in np.loadtxt('Flow_Observed.txt'):
    y.append(num)
    
for num in np.loadtxt('Recon_Observed.txt'):
    z.append(num)
    
# Linear Regression of Observed Data Inputted Into Pandas Table

slope1, intercept1, r_value1, p_value1, std_err1 = linregress(x, y)

table1 = {'Slope':slope1, 'Intercept':intercept1, 'R-Value':r_value1, 'P-Value':p_value1, 'Std Error':std_err1}
linreg1 = pd.DataFrame(table1, index = [0])


# Linear Regression of Tree Ring Reconstructed Data Inputted Into Pandas Table

slope2, intercept2, r_value2, p_value2, std_err2 = linregress(x, z)
table2 = {'Slope':slope2, 'Intercept':intercept2, 'R-Value':r_value2, 'P-Value':p_value2, 'Std Error':std_err2}
linreg2 = pd.DataFrame(table2, index = [0])

#########################################################################
# Start of First Graph
#########################################################################
    
# Adjust Size and dpi of Graph

fig = plt.figure(figsize=(15,6), dpi=100)

# Graph Title

plt.title("Tree Ring Data From 1906 to 2005")

# X & Y Axis Labels

plt.xlabel("Year")
plt.ylabel("Streamflow of Colorado River (Millions Acre/ft)")

# Limit Graph to 1906-2005 X values

#plt.xlim([1906, 2005])

# Linear Polyfit of Tree Ring Reconstructed Data

fit1 = np.polyfit(x, z, 1)
yp = np.polyval(fit1, x)

# Linear Polyfit of Observed Data

fit2 = np.polyfit(x, y, 1)
zp = np.polyval(fit2, x)

# Plots for Linear Regressions

plt.plot(x, zp, label="Linreg of Observed Data", color='red')
plt.plot(x, yp, label="Linreg of Tree Ring Data", color='blue')

# Plots for Tree Ring and Observed Data

plt.plot(x, z, label = 'Tree Ring Data')
plt.plot(x, y, label = 'Actual Observed Data')
plt.legend()
plt.show()

#########################################################################
# Second Graph Displayed With Scatter
#########################################################################

# Adjust Size and dpi of Graph

fig = plt.figure(figsize=(15,6), dpi=100)

# Limit Graph to 1906-2005 X values

#plt.xlim([1906, 2005])

# Graph Title

plt.title("Tree Ring Data From 1906 to 2005")

# X & Y Axis Labels

plt.xlabel("Year")
plt.ylabel("Streamflow of Colorado River (Millions Acre/ft)")

# Plots for Linear Regressions

plt.plot(x, zp, label="Linreg of Observed Data", color='red')
plt.plot(x, yp, label="Linreg of Tree Ring Data", color='blue')

# Plots for Tree Ring and Observed Data

plt.scatter(x, z, label = 'Tree Ring Data',)
plt.scatter(x, y, label = 'Actual Observed Data',)
plt.legend(loc=3)
plt.show()
```

# Fourier Transform of Streamflow Data from the Colorado River at Lees Ferry (1906-2005)

```python
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
import pandas as pd



# Define Empty List for data

x = [] # Years from 1906 - 2005
y = [] # Actual Measured Data 1906 - 2005
z = [] # Tree Ring Data 1906 - 2005


# Load Data & Append to Empty Lists

for num in np.loadtxt('Years_Observed.txt'):
    x.append(num)
    
for num in np.loadtxt('Flow_Observed.txt'):
    y.append(num)
    
for num in np.loadtxt('Recon_Observed.txt'):
    z.append(num)

    
###############################################################################
# Fourier Transform Graph
###############################################################################
    
# Adjust Size and dpi of Graph

fig = plt.figure(figsize=(15,6), dpi=100)

plt.psd(y, label = 'FTT Observed Data')
plt.psd(z, label = 'FTT Reconstructed Data')

# Graph Title

plt.title("Tree Ring Data From 1906 to 2005")

# X & Y Axis Labels

plt.xlabel("Year")
plt.ylabel("Streamflow of Colorado River (Millions Acre/ft)")

plt.legend()

###############################################################################
# Original Graph
###############################################################################

# Adjust Size and dpi of Graph

fig = plt.figure(figsize=(15,6), dpi=100)

plt.plot(x, y, label = 'Observed Streamflow Data')
plt.plot(x, z, label = 'Tree Ring Data')

# Graph Title

plt.title("Tree Ring Data From 1906 to 2005")

# X & Y Axis Labels

plt.xlabel("Year")
plt.ylabel("Streamflow of Colorado River (Millions Acre/ft)")

plt.legend()


plt.show()
```

```python
# Linear Regression Data of Observed FTT

slope7, intercept7, r_value7, p_value7, std_err7 = linregress(plt.psd(y))
table7 = {'Slope':slope7, 'Intercept':intercept7, 'R-Value':r_value7, 'P-Value':p_value7, 'Std Error':std_err7}
linreg7 = pd.DataFrame(table7, index = [0])    
```
